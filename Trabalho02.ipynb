```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import nxviz as nv
import seaborn as sns
import pandas as pd
from nxviz import CircosPlot 
```

Importação arquivo com os aeropostos

```python
airports = pd.read_csv('/content/drive/MyDrive/Cursos/BTI/13 Semestre/Analise de Redes/trabalhos/airports.csv')
```

Importação arquivo com os vous

```python
flights = pd.read_csv('/content/drive/MyDrive/Cursos/BTI/13 Semestre/Analise de Redes/trabalhos/anac.zip')
```

```python
G = nx.Graph()
```

```python
lista = []
```

Gera uma lista dos aeroportos Brasileiros

```python
for index, row in airports.iterrows():
    if row['country'] == 'BRASIL':
        G.add_node(row['code'],
                   region=row['region'],
                   name=row['name']
                  )
        lista.append(row['code']) 
```

```python
print(G.nodes)
```

Adiciona as arestas

```python
for index, row in flights.iterrows():

  if row['origin_airport_abbreviation'] == row['destination_airport_abbreviation']:
        continue
  try:
    fo = lista.index(row['origin_airport_abbreviation'])
    fd = lista.index(row['origin_airport_abbreviation'])
  except:
    fo = -1
    fd = -1
  
  if (fo > -1) and (fd > -1 ):
    G.add_edge(row['origin_airport_abbreviation'], row['destination_airport_abbreviation'])
```

```python
G.edges()
```

```python
print("A rede possue {} nós e {} arestas".format(G.number_of_nodes(),
                                                     G.number_of_edges()))
```

Cria o Circos Plot por Região

```python
c = nv.CircosPlot(G,
                  node_color="region",
                  node_grouping="region",
                  node_order="region",
                  group_order="region",
                  group_legend=False,
                  node_labels=False,
                  group_label_position="region",
                  group_label_color=False,figsize=(10,8))

c.draw()

c.sm.colorbar.remove()

plt.savefig("assortativity.png", 
            format="png",
            dpi=400, 
            bbox_inches="tight",
            transparent=True)


plt.show()
```

Coeficiente por região

```python
nx.attribute_assortativity_coefficient(G,"region")
```

Matrix de mistura

```python
e = nx.attribute_mixing_matrix(G,'region')
e
```

Soma de todos os valores da matriz

```python
np.sum(e)
```

```python
a = np.sum(e, axis=1)
b = np.sum(e, axis=0)

display(a)
display(b)
```

```python
tr_e = np.trace(e)
tr_e
```

Soma da matriz

```python
multi_e = np.sum(np.matmul(e,e))
```

Métrica de variedade 

```python
r = (tr_e - multi_e)/(1 - multi_e)
r
```

Verificando se a Métrica de variedade é igual ao coeficiente por região.

```python
assert (r == nx.attribute_assortativity_coefficient(G,"region"))
```

Grau médio de vizinhos 

```python
degree, avg_neigh_degree = zip(*nx.average_degree_connectivity(G).items())
```

```python
degree = list(degree)
avg_neigh_degree = list(avg_neigh_degree)
```

```python
plt.style.use("fivethirtyeight")
fig, ax = plt.subplots(1,1,figsize=(12,8))

sns.regplot(degree,avg_neigh_degree,ax=ax)

ax.set_xlabel("Grau do Nó")
ax.set_ylabel("Grau médio de vizinhos")
ax.set_xlim(0,65)


plt.show()
```
