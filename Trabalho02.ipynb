```python
import networkx as nx
import matplotlib as plt
import numpy as np
import nxviz as nv
import seaborn as sns
import pandas as pd
```

Importação arquivo com os aeropostos

```python
airports = pd.read_csv('/content/drive/MyDrive/Cursos/BTI/13 Semestre/Analise de Redes/trabalhos/airports.csv')
```

Importação arquivo com os vous

```python
flights = pd.read_csv('/content/drive/MyDrive/Cursos/BTI/13 Semestre/Analise de Redes/trabalhos/anac.zip')
```

```python
G = nx.Graph()
```

```python
lista = []
```

Gera uma lista dos aeroportos Brasileiros

```python
for index, row in airports.iterrows():
    if row['country'] == 'BRASIL':
        G.add_node(row['code'],
                region=row['region']
                )
        lista.append(row['code']) 
```

```python
print(G.nodes)
```

Adiciona as arestas

```python
for index, row in flights.iterrows():

  if row['origin_airport_abbreviation'] == row['destination_airport_abbreviation']:
        continue
  try:
    fo = lista.index(row['origin_airport_abbreviation'])
    fd = lista.index(row['origin_airport_abbreviation'])
  except:
    fo = -1
    fd = -1
  
  if (fo > -1) and (fd > -1 ):
    G.add_edge(row['origin_airport_abbreviation'], row['destination_airport_abbreviation'])
```

```python
G.edges()
```

```python
print("The network has {} nodes and {} edges".format(G.number_of_nodes(),
                                                     G.number_of_edges()))
```

```python
G.nodes(data=True)
```

Cria o Circos Plot por Região

```python
c = nv.CircosPlot(G,
                  node_color="region",
                  node_grouping="region",
                  node_order="region",
                  group_order="alphabetically",
                  group_legend=False,
                  node_labels=False,
                  group_label_position="middle",
                  group_label_color=False,figsize=(10,8))

c.draw()

c.sm.colorbar.remove()

plt.savefig("assortativity.png", 
            format="png",
            dpi=400, 
            bbox_inches="tight",
            transparent=True)


plt.show()
```
